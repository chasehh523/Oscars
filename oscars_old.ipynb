```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor, MLPClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("BestPictureNominations.csv")
data=data.drop(labels=['Win', 'SAG', 'WGA', 'DGA', 'PGA', 'BAFTA', 'OscarNoms', 'DirectingNom', 'ActingNom', 'ScreenplayNom', 'EditingNom'], axis=1)
#data=data[data['Year']!=2023]
train=data[data['Year']!=2020]
y_train=train['Best Picture Nominee']
X_train=train.drop(labels=['Best Picture Nominee', 'Title', "Year"], axis=1)

test=data[data['Year']==2020]
y_test=test['Best Picture Nominee']
X_test_titles=test['Title']
X_test=test.drop(labels=['Best Picture Nominee', "Title", "Year"], axis=1)

scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
log_r = LogisticRegression()
#log_r.fit(X_train, y_train)
#log_r=SelectFromModel(log_r, prefit=True)
#X_train=log_r.transform(X_train)
#X_test=log_r.transform(X_test)
#log_r=LogisticRegression()
log_r.fit(X_train, y_train)
y_pred1 = log_r.predict_proba(X_test)
#print(X_test_titles)
#y_pred
output=pd.DataFrame()
output['Title']=X_test_titles
output['Probability']=y_pred1[:,1]
output.sort_values(by=['Probability'], axis=0, ascending=False)[:12]
```

```python
rfr = RandomForestRegressor(n_estimators=200)
rfr.fit(X_train, y_train)
y_pred2 = rfr.predict(X_test)

#print(X_test_titles)
#print(y_pred)
output=pd.DataFrame()
output['Title']=X_test_titles
output['Probability']=y_pred2
output.sort_values(by=['Probability'], axis=0, ascending=False)[:12]
```

```python
#log_r = LogisticRegression()
#log_r.fit(X_train, y_train)
#log_r=SelectFromModel(log_r, prefit=True)
#X_train=log_r.transform(X_train)
#X_test=log_r.transform(X_test)
nn = MLPRegressor(hidden_layer_sizes=(200,), activation='logistic')
nn.fit(X_train, y_train)
y_pred3 = nn.predict(X_test)


#print(X_test_titles)
#print(y_pred)
output=pd.DataFrame()
output['Title']=X_test_titles
output['Probability']=[round(100*min(i, 1),2) for i in y_pred3]
output.sort_values(by=['Probability'], axis=0, ascending=False)
#print(output.loc[41])
```

```python
y_pred_ensemble=(y_pred1[:,1]+y_pred2+y_pred3)/3
output=pd.DataFrame()
output['Title']=X_test_titles
output['Probability']=y_pred_ensemble
output.sort_values(by=['Probability'], axis=0, ascending=False)[:12]
print()
```
